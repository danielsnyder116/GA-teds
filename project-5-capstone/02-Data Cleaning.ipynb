```python
nordeste = pd.read_csv(path, encoding="latin1")

pima = nordeste.loc[nordeste["SG_UF_RESIDENCIA"].isin(["PI", "MA"]), :]

new_columns = ["student_id_num", "city_of_residence", "state_of_residence_initial_convert_to_name", "age", "sex", 
               "marital_status", "race", "nationality", "birth_city_code_drop?", "birth_city_name", "birth_city_state_number_DROP", 
               "birth_state_initials", "student_hs_grad_status", "high_school_graduation_year", "high_school_type", 
               "hs_institution_type_maybe_call_hs_format", "for_fun?",  
               
               "school_code", "number_generated_to_align_with_education_census_DROP", "name_of_school_city", "state_code_DROP", 
               "state_initials_convert_to_name", "school_admin_level_fedvsstatevslocalvsprivate", "school_location_urbvs.rur", 
               "school_status_still_running?", 

               "low_vision?", "blind?", "deaf?", "hearing_deficiency?", "blind_and_deaf?", 
               "physical_deficiency?", "mental_deficiency?", "ADD?_attention_deficit", "dyslexic?", 
               "dsycalculia?_arith_trouble", "autism?", "monocular_vision?", "other_condition_special_need?", 
            
               "pregnant?", "lactating?", "elderly?", "admitted_to_hospital?", 
               
               "requested_resource?",
               "braille?", "enlarged_font24?", "enlarged_font18?", "reading_aid_reader?", "easy_access_room?", 
               "transcription_request?", "translation_libras_interpreter?", "lip_reading_aid?", "wheel_chair_or_handicap_desk?", 
               "separated_desk_and_chair?", "leg_or_foot_support?", "guide_interpreter?_drop?", "computer?", "special_chair?", 
               "left_handed_chair?", "padded_chair?", "took_test_lieing_down_or_on_stretcher_or_furniture?", 
               "furniture_for_obese?", "overlay_blade?", "protective_headphones_headset?", "glucose_measurer_or_insulin_request?", 
               "Braille_machine_or_ruler/puncture?", "soroban?japanese_abacus", "has_pace_maker?_avoids_metal_inspection", 
               "periodically_changed_stylet?", "medication_request?", "individual_room?","special_room_up_to_20?", 
               "reserved_room_for_escort/companion?","specific_furniture?", "specific_material?", 
               "transgender?", 
               
               "city_name_where_exam", "state_initials_where_exam_convert_to_name", 
               
               "did_natural_science_part?","did_human_sciences_part?", "did_language_and_code_part?", "did_math_part?", 
               "nat_science_test_type", "human_science_test_type", "lang_code_test_type", "math_test_type", "ns_score", "hs_score", 
               "lc_score", "math_score","ns_response_vector", "hs_response_vector", "lc_response_vector", "math_response_vector", 
               "for_language_choice","ns_answer_key", "hs_answer_key", "lc_answer_key", "math_answer_key", 
               
               "essay_gradable?", "competence1_score",
               "competence2_score", "competence3_score", "competence4_score", "competence5_score", "essay_score", 
               
               "dad_male_guardian_education", "mom_female_guardian_education", "male_occupation_class", "female_occupation_class",
               "number_of_housemates_include_student", "family_monthly_income", "domestic_worker?", "home_has_bathroom?", 
               "home_has_bedrooms?", "home_has_car?", "home_has_scooter/motorcycle?", "home_has_fridge?", "home_has_freezer?", 
               "home_wash_machine?","home_has_dryer?","home_has_microwave?", "home_has_dishwasher?", "home_has_vacuum?", 
               "home_has_color_tv?","home_has_DVD_player?", "home_has_cable/subscribed_tv?", "home_has_cellphone?", "home_has_landline_phone?",
               "home_computer", "home_has_internet?", "student_high_school_status_again", "high_school_type_and_scholarship?"]
```

```python
pima.drop(["braille?", "enlarged_font24?", "enlarged_font18?", "reading_aid_reader?", "easy_access_room?", 
             "transcription_request?", "translation_libras_interpreter?", "lip_reading_aid?", "wheel_chair_or_handicap_desk?", 
               "separated_desk_and_chair?", "leg_or_foot_support?", "guide_interpreter?_drop?", "computer?", "special_chair?", 
               "left_handed_chair?", "padded_chair?", "took_test_lieing_down_or_on_stretcher_or_furniture?", 
               "furniture_for_obese?", "overlay_blade?", "protective_headphones_headset?", "glucose_measurer_or_insulin_request?", 
               "Braille_machine_or_ruler/puncture?", "soroban?japanese_abacus", "has_pace_maker?_avoids_metal_inspection", 
               "periodically_changed_stylet?", "medication_request?", "individual_room?","special_room_up_to_20?", 
               "reserved_room_for_escort/companion?","specific_furniture?", "specific_material?", ], axis=1, inplace=True)
```

```python
pima.drop(["competence1_score", "competence2_score", "competence3_score", "competence4_score", "competence5_score"], axis=1,inplace=True)
pima.drop(["home_has_vacuum?", "home_has_DVD_player?", "home_has_cable/subscribed_tv?", "home_has_cellphone?",
         "home_has_landline_phone?", "admitted_to_hospital?", "city_name_where_exam", 
           "state_initials_where_exam_convert_to_name", "student_high_school_status_again"], axis=1, inplace=True)

pima.drop(["hs_institution_type_maybe_call_hs_format", "school_admin_level_fedvsstatevslocalvsprivate",
          "school_location_urbvs.rur", "school_status_still_running?"], axis=1, inplace=True)

pima.drop(["nat_science_test_type", "human_science_test_type", "lang_code_test_type", "math_test_type"], axis=1, inplace=True)

pima.drop(["lactating?", "elderly?"], axis=1, inplace=True)

pima.drop(["home_has_bedrooms?","home_has_car?","home_has_scooter/motorcycle?", "home_has_fridge?",
          "home_has_freezer?", "home_has_dryer?", "home_has_microwave?","home_has_dishwasher?",
         "home_has_color_tv?", "high_school_type_and_scholarship?"], axis=1, inplace=True)
```

```python
pima["state_of_residence_initial_convert_to_name"] = pima["state_of_residence_initial_convert_to_name"].str.replace("PI", "Piauí")
pima["state_of_residence_initial_convert_to_name"] = pima["state_of_residence_initial_convert_to_name"].str.replace("MA", "Maranhão")
```

```python
pima.rename(columns={"state_of_residence_initial_convert_to_name": "state_of_residence"}, inplace=True)
pima.rename(columns={"did_natural_science_part?":"did_ns_math?", "did_human_sciences_part?":"did_hs_lang?"}, inplace=True)
pima.drop(["did_language_and_code_part?", "did_math_part?"], axis=1, inplace=True)
```

```python
pima["family_monthly_income"] = pima["family_monthly_income"].map({"A":"no_response", "B":"0-937", "C":"937-1406", "D":"1406-1874",
     "E":"1874-2343","F":"2343-2811", "G":"2811-3748", "H":"3748-4685", "I":"4685-5622", "J":"5622-6559", 
     "K":"6559-7496", "L":"7496-8433","M":"8433-9370", "N":"9730-11244", "O":"11244-14055", "P":"14055-18740", "Q":"18740+"})
```

```python
pima["income_avr"] = pima["family_monthly_income"].map({"no_response":0, "0-937":(937/2), "937-1406":(1406+937)/2, 
        "1406-1874":(1406+1874)/2,"1874-2343":(1874+2343)/2, "2343-2811":(2343+2811)/2, "2811-3748":(2811+3748)/2, 
        "3748-4685":(3748+4685)/2, "4685-5622":(4685+5622)/2, "5622-6559": (5622+6559)/2, "6559-7496":(6559+7497)/2, 
        "7496-8433":(7496+8433)/2, "8433-9370":(8433+9370)/2, "9730-11244":(9730+11244)/2, "11244-14055": (11244+14055)/2, 
        "14055-18740":(14055+18740)/2, "18740+":18740 })
```

```python
pima["ns_score"].fillna(value=0, inplace=True)
pima["hs_score"].fillna(value=0, inplace=True)
pima["lc_score"].fillna(value=0, inplace=True)
pima["math_score"].fillna(value=0, inplace=True)
pima["for_language_choice"].fillna(value=0, inplace=True)
pima["essay_score"].fillna(value=0, inplace=True)
```

```python
## Renaming columns, forming new columns, and mapping out elements.

pima["dad_male_guardian_education"] = pima["dad_male_guardian_education"].map({"A":"no_education", "B":"fifth_grade",
    "C":"between_5th_9th", "D":"completed_9th", "E":"high_school","F":"college", "G":"graduate_degree", "H":"don't_know"})

pima["mom_female_guardian_education"] = pima["mom_female_guardian_education"].map({"A":"no_education", "B":"fifth_grade",
     "C":"between_5th_9th", "D":"completed_9th", "E":"high_school", "F":"college", "G":"graduate_degree", "H":"don't_know"})

pima["male_occupation_class"] = pima["male_occupation_class"].map({"A":"farmer1", "B":"unskilled_blue2", "C":"skilled_blue3",
     "D":"low_white4","E":"skilled_white5", "F":"don't_know"})

pima["high_school_graduation_year"] = pima["high_school_graduation_year"].map(
    {0:"no_response", 1:"2016", 2:"2015", 3:"2014", 4:"2013", 5:"2012", 6:"2011", 7:"2010", 
     8:"2009", 9:"2008", 10:"2007", 11:"before_2007"})

pima["female_occupation_class"] = pima["female_occupation_class"].map({"A":"farmer1", "B":"unskilled_blue2", "C":"skilled_blue3",
     "D":"low_white4","E":"skilled_white5", "F":"don't_know"})

pima["race"] = pima["race"].map({0:"no_response", 1:"White", 2:"Black", 3:"Mestizo", 4:"Asian", 5:"Native"})

pima["nationality"] = pima["nationality"].map({0:"no_response", 1:"brazilian", 2:"naturalized", 3:"foreigner", 4:"native_born_abroad"})

pima["per_house_member_income"] = (pima["income_avr"])/(pima["number_of_housemates_include_student"])

pima["home_has_internet?"] = pima["home_has_internet?"].map({"A":0, "B":1})

pima["domestic_worker?"] = pima["domestic_worker?"].map({"A":"No", "B":"Yes", "C":"Yes", "D":"Yes"})
       # In this case, if you have someone coming to clean your house, that is what matters - mapping to binary. 
    
pima["total_score"] = (pima["ns_score"] + pima["hs_score"] + pima["lc_score"] + pima["math_score"] + pima["essay_score"])

pima["marital_status"] = pima["marital_status"].map({0:"Single", 1:"Married", 2:"Divorced", 3:"Widow(er)"})

pima["student_hs_grad_status"] = pima["student_hs_grad_status"].map({1:"finished", 2:"finishing_in_2017",
                                   3:"finishing_after_2017", 4:"did_not_finish_not_enrolled"})

pima["high_school_type"] = pima["high_school_type"].map({1:"no_response", 2:"public", 3:"private", 4:"foreign"})
pima.rename(columns={"high_school_type": "hs_type"}, inplace=True)

pima["for_fun?"] = pima["for_fun?"].map({0:"no,_student", 1:"yes_not_student"})

pima["did_ns_math?"] = pima["did_ns_math?"].map({0:"Absent", 1:"Present"})
     
pima["did_hs_lang?"] = pima["did_hs_lang?"].map({0:"Absent", 1:"Present"})

pima["essay_gradable?"] = pima["essay_gradable?"].map({0:"no_response", 1:"all_good", 2:"nullified", 3:"copy of prompt", 
                                                       4:"blank", 5:"?", 6:"doesn't_respond_to_prompt", 7:"no_attention_to_type",
                                                       8:"not_enough_text", 9:"disconnected_part"})

pima["home_has_internet?"] = pima["home_has_internet?"].map({0:"No", 1:"Yes"})

pima["home_computer"] = pima["home_computer"].map({"A":0, "B":1, "C":2,"D":3, "E":4})

pima["home_wash_machine?"] = pima["home_wash_machine?"].map({"A":"No", "B":"Yes", "C":"Yes","D":"Yes", "E":"Yes"})

pima["home_has_bathroom?"] = pima["home_has_bathroom?"].map({"A":"No", "B":"Yes", "C":"Yes", "D":"Yes", "E":"Yes"})

pima["for_language_choice"] = pima["for_language_choice"].map({0:"English", 1:"Spanish"})
```

```python
pima.isnull().mean()
```

```python
### Filling in nan values for columns with low percentages of missing data.

pima["marital_status"].fillna(value="Missing", inplace=True)
pima["birth_city_name"].fillna(value="Missing", inplace=True)                                     
pima["birth_state_initials"].fillna(value="Missing", inplace=True)
pima["age"].fillna(value=0, inplace=True)
pima["total_score"].fillna(value=0, inplace=True)
pima["did_ns_math?"].fillna(value="Missing", inplace=True)
pima["did_hs_lang?"].fillna(value="Missing", inplace=True)
pima["essay_gradable?"].fillna(value="Missing", inplace=True)
```
